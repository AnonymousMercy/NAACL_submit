```python
import json
from tqdm import tqdm
import os

directory_path = "/data02/MLP/naacl/LLaVA/utils/short"
directories = os.listdir(directory_path)

for file_name in directories:
    output_path=f'./short/dot/{file_name}'
    output_list = []
    if file_name.endswith('.json'):
        print(file_name)
        with open(directory_path + '/' + file_name, 'r') as file:
            data = json.load(file)
            for line in tqdm(data):
                # print(line['answer'])
                outputs = line['answer'].split('.')[0]
                outputs += '.'
                line['answer'] = outputs
                output_list.append(line)
    with open(output_path, 'w') as outputdir:
        json.dump(output_list, outputdir, indent=4, ensure_ascii=False)


# with open('./short/[MLP-KTLim-X_LLaVA_O_notconver_BaseLLM_L_short_30]_[coco2014_val_qa_eval]_[ko].json', 'r') as file:
#     data = json.load(file)
#     for line in tqdm(data):
#         # print(line['answer'])
#         outputs = line['answer'].split('.', 1)[0]
#         outputs += '.'
#         line['answer'] = outputs
#         output_list.append(line)

# with open(output_path, 'w') as file:
#     json.dump(output_list, file, indent=4, ensure_ascii=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 샘플 데이터 생성
data = {
    'XLLaVA Wins':  [15,  64, 54, 72,  82],
    'Tie':          [37,  24, 33, 16,   8],
    'XLLaVA Loses': [38,   2,  3,  2,   0]
}

df = pd.DataFrame(data)
df_percentage = df.divide(df.sum(axis=1), axis=0) * 100

fig, ax1 = plt.subplots(figsize=(20, 14))

colors = ['#272829', '#61677A',  '#D8D9DA']
bars = df_percentage.plot(kind='barh', stacked=True, color=colors, ax=ax1)

# ax1.set_yticklabels(['XLLaVA'] * 6, horizontalalignment='right', weight='bold')
# ax1.set_yticklabels(['XLLaVA'] * len(data['XLLaVA Loses']), horizontalalignment='right', weight='bold')

# 왼쪽 숫자 없애기
ax1.set_yticklabels([])

ax2 = ax1.twinx()
ax2.set_ylim(ax1.get_ylim())
ax2.set_yticks(ax1.get_yticks())
# ax2.set_yticklabels(['GPT4-VISION', 'KoLLaVA', 'LLaVA-v1.5', 'InstructBLIP', 'BLIP2'], horizontalalignment='left', weight='bold')
ax2.set_yticklabels(['GPT4-VISION', 'KoLLaVA', 'LLaVA-v1.5', 'InstructBLIP', 'BLIP2'], horizontalalignment='left', weight='bold')


# 텍스트를 추가하는 부분
for idx, (name, row) in enumerate(df_percentage.iterrows()):
    for col, value in enumerate(row):

        if value <= 2:
            continue
        if (col == 0) or (col == 1):
            width = bars.containers[col][idx].get_width()
            height = bars.containers[col][idx].get_height()
            x = bars.containers[col][idx].get_x() + width / 2
            y = bars.containers[col][idx].get_y() + height / 2
            ax1.text(x, y, f'{value:.0f}%', ha='center', va='center', color='white', fontsize=32)
            ax1.set_xlim(0, 100)
        else:
            width = bars.containers[col][idx].get_width()
            height = bars.containers[col][idx].get_height()
            x = bars.containers[col][idx].get_x() + width / 2
            y = bars.containers[col][idx].get_y() + height / 2
            ax1.text(x, y, f'{value:.0f}%', ha='center', va='center', color='black', fontsize=32)

ax1.set_xlim(0, 100)
ax1.set_xticks(list(range(0, 101, 20)))
ax1.set_xticklabels([f'{x}%' for x in range(100, -1, -20)])
ax1.tick_params(axis='x', labelsize=25)
ax1.tick_params(axis='y', labelsize=30)
ax1.get_legend().remove()

# 이중 x축 생성
ax3 = ax1.twiny()
ax3.set_xlim(ax1.get_xlim())
ax3.set_xticks(ax1.get_xticks())
ax3.tick_params(axis='x', labelsize=25)
ax3.set_xticklabels([f'{x}%' for x in range(0, 101, 20)])
ax2.tick_params(axis='y', labelsize=30)
ax1.grid(False)
ax2.grid(False)
ax3.grid(False)
ax1.legend(bbox_to_anchor=(0.5 , 1.15), fontsize = 30, loc='upper center', ncol=3, frameon = False)
# plt.savefig("/home/hslim/LLM/colling_graph/Bllossom_vs_kor_llm.png", dpi=600, bbox_inches='tight')

plt.savefig("/data02/MLP/naacl/LLaVA/utils/figs/XLLaVA_vs_Others_short - KO.pdf", bbox_inches='tight')
plt.savefig("/data02/MLP/naacl/LLaVA/utils/figs/XLLaVA_vs_Others_short - KO.png", bbox_inches='tight')
plt.show()
```
